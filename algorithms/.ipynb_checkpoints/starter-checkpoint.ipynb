```python
import matplotlib.pyplot as plt
import numpy as np
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram
from numpy.random import randint
import pandas as pd
print("Imports Successful")
```

```python
circuit = QuantumCircuit(4,2) # indicate number of qubits and classical bits
```

```python
circuit.draw(output='mpl')
```

```python
circuit.x(0) # Applying an X gate to 0-th qubit
circuit.x(1) # 
circuit.barrier() # just for a visual separation
```

```python
circuit.draw(output='mpl')
```

```python
circuit.cx(0,2) # CNOT gate that is controlled by 0 and targets 2
```

```python
circuit.draw(output="mpl")
```

```python
circuit.cx(1,2)
circuit.ccx(0,1,3) # use ccx (C-C-NOT) to write the AND of the inputs on qubit 3
circuit.barrier()
```

```python
circuit.draw(output="mpl")
```

```python
circuit.measure(2,0) # extract XOR value. Measure 2-nd qubit to 0-th classical bit
circuit.measure(3,1) # extract AND value. Measure 3-rd qubit to 1-th classical bit
circuit.barrier()
```

```python
circuit.draw(output="mpl")
```

```python
job = assemble(circuit)
```

```python
backend = Aer.get_backend('aer_simulator')  # Getting a simulator 
```

```python
counts = backend.run(job).result().get_counts()
plot_histogram(counts)
```
